<h3>Part 2-6: Create a function that reads the wanted information, then displays the data in a table</h3>

```python
import pandas as pd
import numpy as np
from IPython.display import display

def table(rows=-1,file="data/2021-09-14_party distribution_1_st_2021.csv"):

    # read data, obly keep relevant columns
    parties = pd.read_csv(file,
                          sep=";",
                          encoding="utf-8",
                          usecols=["Partikode", 
                                   "Antall stemmer totalt"])
    
    # group by party code and sum votes 
    parties = parties.groupby(["Partikode"]).sum()
    
    # reset the index to make the party code a column
    parties.reset_index(inplace=True)

    parties.rename(columns={"Antall stemmer totalt":"Votes"},inplace=True)

    # calculate percentages
    votes = parties["Votes"].to_numpy()
    percentages = np.round(100*votes/np.sum(votes),2) # round vector to 2 decimals
    
    # create new coloumns
    condition = percentages>4
    above_four_percent = np.array(["**" if i else " " for i in condition])

    parties["percentages"] = percentages
    parties["Above 4%"] = above_four_percent # vector of booleans


    # Make sure variable rows has a valid value
    if rows not in range(len(parties)): rows=len(parties)
    
    # print sorted table with given number of rows
    display(parties.sort_values(by=["Votes"], ascending=False).iloc[0:rows])

table(6) # table(7) table()
```

Task 2: Regular expressions

```python
import re

sentences =["Ali and Per and friends.",
            "Kari and Joe know each other.",
            "James has known Peter since school days."]

def find_pairs(sentence):
    return re.findall(r"(^[A-Z][a-z]+).+\b([A-Z][a-z]+)", sentence)

print(f"{'Friendships':^25}\n{'-'*25}")	
for sentence in sentences:
    result = find_pairs(sentence)
    try: 
        left, right = result[0]
        print(f"{left:>11} - {right:<11}")
    except:
        print(f"{sentence} contains no pairs")
    
    
```

<p>
\b matches word boundary, in this case the beginning of a word
</p>
<p>
[A-Z] matches any capital letter/n
</p>
<p>
\w* matches any number of word characters, lower or upper case.\n
</p>
<p>
the pattern is thus: any word starting with a capital letter\n
</p>
<p>
followed any number of characters, except spaces.\n
</p>
