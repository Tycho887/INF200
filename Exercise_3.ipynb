<h1>Task 1: reading files and analysing data</h1>

```python
with open("data/students.txt") as f:
    students = []
    for line in f: 
        name, num_vals = line.rstrip('\n').split(':')
        age, phone_number = num_vals.split(',')
        students.append({"Name":name,"Age":age,"Phone number":phone_number})
    print(students)
```

<h1>Task 2: Navigating directories, and regular expressions</h1>


```python
import re
from pathlib import Path

cwd = Path.cwd()

def get_packages(file):
    with open(file) as f:
        pattern = re.compile(r"(?:from|import)\s(\w+)")
        return pattern.findall(f.read())

def print_results(path):
    print(f"In {path.name} we found the following packages:")
    for package in get_packages(path): 
        print(f"\t{package}")

def search_directory():
    for path in cwd.glob("../**/*.py"):
        if path.is_file() and "conda" not in str(path):
            try: print_results(path)
            except: pass
               
search_directory()
```
