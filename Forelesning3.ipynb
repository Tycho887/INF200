<h3>What is a path? A way to navigate through a file system</h3>

Each path has:
1. A staring point
2. Steps

Absolute path
1. Starts from the root directory



```python
from pathlib import Path

cwd = Path.cwd() # Path variable

file_name = "main.py"
file_path = cwd / file_name # we use / to add subpaths

# An absolute path is a path that starts from the root of the file system

file_path = file_path.resolve() # resolve() returns the absolute path
print(file_path)

if file_path.is_absolute():
    print("this path is absolute")

print(list(cwd.glob('**/*.py')))
```

```python
from pathlib import Path
import re

NaN_finder = re.compile(r".*NaN.*")

folder = Path.cwd() / "data"

for file in folder.glob('**/*.txt'):
    with open(file) as f:
        for line in f:
            out = NaN_finder.findall(line)
            if len(out) > 0:
                print("There are NaN values in file", file)
                break
        
```
